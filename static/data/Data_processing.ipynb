```python
import pandas as pd
import datetime
```

```python
path = "../Major_Crime_Indicators_Open_Data.csv"

data_df = pd.read_csv(path)
data_df.head()
```

```python
filter_df = data_df[data_df['OCC_YEAR']>2020]
```

```python
filter_df['REPORT_DATE'] = pd.to_datetime(filter_df['REPORT_DATE'])
```

```python

filter_df['REPORT_DATE'] = filter_df['REPORT_DATE'].apply(lambda dt: dt.date())
```

```python
filter_df.head()
```

```python
filter_df['OCC_DATE'] = pd.to_datetime(filter_df['OCC_DATE'])
filter_df['OCC_DATE'] = filter_df['OCC_DATE'].apply(lambda dt: dt.date())
```

```python
filter_df.head()
```

```python
final_df = filter_df[['X', 'Y', 'OBJECTID', 'EVENT_UNIQUE_ID', 'REPORT_DATE', 'OCC_DATE',
        'OCC_YEAR', 'OCC_MONTH','OCC_HOUR', 'DIVISION', 'LOCATION_TYPE', 'PREMISES_TYPE', 'OFFENCE', 'MCI_CATEGORY', 'HOOD_158', 'NEIGHBOURHOOD_158',
       'HOOD_140', 'NEIGHBOURHOOD_140']]
```

```python
def convert_to_date(date_str):
    return pd.to_datetime(date_str, format='%m/%d/%Y')  # Adjust the format as needed

# Apply the function to the 'REPORT_DATE' and 'OCC_DATE' columns
final_df['REPORT_DATE'] = final_df['REPORT_DATE'].apply(convert_to_date)
final_df['OCC_DATE'] = final_df['OCC_DATE'].apply(convert_to_date)
```

```python
final_df = final_df[final_df['Y']!=0]
```

```python
final_df = final_df[final_df['HOOD_140']!='NSA']
```

```python
final_df.head()
```

```python
final_df.reset_index()
```

```python
final_df['OBJECTID'] = final_df.index
```

```python
import csv
import json
input_file = 'Modified_major_crime.csv'
output_file = 'Modified_major_crime.json'

csv_records = []
with open(input_file, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        csv_records.append(row)

# Write JSON with commas between records
with open(output_file, 'w') as json_file:
    json_file.write('[')
    for index, record in enumerate(csv_records):
        json.dump(record, json_file, indent=2)
        if index < len(csv_records) - 1:
            json_file.write(',')
    json_file.write(']')
```
